```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (ensure this path is correct)
df = pd.read_csv('/Users/fadumom/Downloads/healthcare_prices.csv')

# Display the first few rows and columns of the DataFrame
print(df.head())
print(df.columns)

# Rename the columns if necessary (to remove extra spaces or special characters)
df.columns = df.columns.str.strip()  # This removes any leading or trailing whitespace

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, 
                 x='Healthcare prices, 2017',  # Change to your actual column name for X-axis
                 y='Global rank',                # Change to your actual column name for Y-axis
                 hue='Countries',                # You can choose to use 'Countries' or any categorical column for color
                 style='Countries',              # Optional: Use this if you want different styles for different countries
                 s=100)                          # Size of the markers
plt.title('Healthcare Prices vs Global Rank (2017)')
plt.xlabel('Healthcare Prices (2017)')
plt.ylabel('Global Rank')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend placement if necessary
plt.show()
```

```python
print(df.dtypes)
```

```python
df.dropna(subset=['Healthcare prices, 2017', 'Global rank'], inplace=True)
```

```python
# Histogram of healthcare prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Healthcare prices, 2017'], bins=30, kde=True)
plt.title('Distribution of Healthcare Prices (2017)')
plt.xlabel('Healthcare Prices (2017)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a bar plot for the top 10 countries with the highest healthcare prices
top_10 = df.nlargest(10, 'Healthcare prices, 2017')

plt.figure(figsize=(12, 6))
sns.barplot(data=top_10, 
            x='Countries', 
            y='Healthcare prices, 2017', 
            palette='viridis')
plt.title('Top 10 Countries by Healthcare Prices (2017)')
plt.xlabel('Countries')
plt.ylabel('Healthcare Prices (2017)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram of healthcare prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Healthcare prices, 2017'], bins=30, kde=True, color='blue')
plt.title('Distribution of Healthcare Prices (2017)')
plt.xlabel('Healthcare Prices (2017)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Box plot of healthcare prices
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Healthcare prices, 2017')
plt.title('Box Plot of Healthcare Prices (2017)')
plt.xlabel('Healthcare Prices (2017)')
plt.grid(True)
plt.show()
```

```python
# Create a correlation heatmap
corr_data = df[['Healthcare prices, 2017', 'Global rank']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_data, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
pair_plot = sns.pairplot(df[['Healthcare prices, 2017', 'Global rank']], height=3)
plt.suptitle('Pair Plot of Healthcare Prices and Global Rank', y=1.02)
plt.show()
```
